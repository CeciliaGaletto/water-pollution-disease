** Abstract **

Este conjunto de datos examina la relación entre la contaminación del agua y la prevalencia de enfermedades de transmisión hídrica a nivel mundial, ofreciendo una visión integral que abarca tanto aspectos ambientales como sociales. Incluye indicadores de calidad del agua, niveles de contaminación, tasas de enfermedades y factores socioeconómicos que influyen directamente en los resultados de salud. La información corresponde a distintos países y regiones y cubre un periodo de veintiséis años, desde 2000 hasta 2025, lo que permite analizar tendencias temporales y comparaciones entre contextos geográficos.

El conjunto de datos aborda variables fundamentales como los niveles de contaminantes químicos y biológicos, el acceso a fuentes de agua potable, la presencia de bacterias y microorganismos nocivos, los métodos de tratamiento aplicados, la cobertura de saneamiento y la incidencia de enfermedades vinculadas al agua, entre ellas la diarrea, el cólera y la fiebre tifoidea. Al integrar estas dimensiones, se facilita la identificación de vínculos causales y patrones epidemiológicos que permiten comprender cómo la contaminación hídrica afecta de manera diferenciada a las comunidades.

Además, se incorporan variables socioeconómicas que enriquecen el análisis, tales como el PBI per cápita, la tasa de urbanización y el acceso a los servicios de salud. Estas dimensiones son esenciales para evaluar de forma más amplia los impactos de la contaminación, pues reflejan cómo las condiciones económicas y sociales influyen en la vulnerabilidad de la población y en su capacidad de respuesta frente a riesgos sanitarios.

Este conjunto de datos tiene múltiples aplicaciones: puede emplearse en investigaciones en salud pública sobre los efectos de la contaminación del agua; en estudios ambientales para identificar tendencias en la calidad de los recursos hídricos; en la formulación de políticas públicas orientadas a garantizar el acceso universal a agua limpia y a mejorar las infraestructuras de saneamiento; e incluso en el desarrollo de modelos de aprendizaje automático capaces de predecir brotes de enfermedades en función de los indicadores de calidad del agua.

```python
#Normalizar nombre de las columnas
import unicodedata, re

def normalize_cols(cols):
    out = []
    for c in cols:
        # elimina acentos
        c = unicodedata.normalize('NFKD', str(c)).encode('ascii','ignore').decode('ascii')
        # reemplaza espacios y símbolos por guiones bajos
        c = re.sub(r'[^A-Za-z0-9]+', '_', c.strip().lower()).strip('_')
        out.append(c)
    return out
```

```python
df.columns = normalize_cols(df.columns)
df.head(2)
```

```python
# Variables categóricas
cat_candidates = ['country','region','water_source_type','water_treatment_method']
for c in cat_candidates:
    if c in df.columns:
        df[c] = df[c].astype('category')

# Año como numérico + versión categórica
if 'year' in df.columns:
    df['year'] = pd.to_numeric(df['year'], errors='coerce').astype('Int64')
    df['year_cat'] = df['year'].astype('string')  # útil para gráficos

# Variables numéricas
num_candidates = [
    'contaminant_level_ppm','ph_level','turbidity_ntu','nitrate_level_mg_l','bacteria_count_cfu_ml',
    'access_to_clean_water_of_population','diarrheal_cases_per_100_000_people','cholera_cases_per_100_000_people',
    'typhoid_cases_per_100_000_people',
    'infant_mortality_rate_per_1_000_live_births','healthcare_access_index_0_100',
    'urbanization_rate','temperature'
]
for c in num_candidates:
    if c in df.columns:
        df[c] = pd.to_numeric(df[c], errors='coerce')

# chequeo rápido
print(df.info())
df.head(3)
```

Para avanzar en el proyecto, resulta fundamental plantear algunas preguntas e hipótesis que guiarán el análisis del dataset. Dado que los datos disponibles combinan información sobre la calidad del agua, acceso a recursos básicos y prevalencia de enfermedades relacionadas, se busca indagar principalmente en las relaciones entre contaminación, acceso a servicios y salud pública.

En este sentido, las hipótesis iniciales que se proponen son las siguientes:

**A mayor nivel de contaminación del agua, mayor es la incidencia de enfermedades de origen hídrico, como diarrea, cólera y tifoidea.**

**Diferentes métodos de tratamiento del agua se asocian con distintos niveles de contaminación, de manera que un tratamiento más efectivo reduce los niveles de contaminación.**

**La mortalidad infantil está influida tanto por la calidad del agua como por el acceso a la salud, siendo estos dos factores determinantes en la prevención de muertes evitables.**

A partir de estas hipótesis, el análisis exploratorio se centrará en realizar visualizaciones y resúmenes numéricos que permitan observar patrones, contrastar tendencias y evaluar si los datos apoyan o contradicen las suposiciones iniciales.

En los apartados siguientes se presentarán gráficos univariados, bivariados y multivariados, junto con interpretaciones que servirán como diagnóstico preliminar para responder las preguntas planteadas.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
#gráfico univariado
plt.figure(figsize=(8,5))
sns.histplot(df['contaminant_level_ppm'], kde=True, color='skyblue')
plt.title("Distribución del nivel de contaminación")
plt.xlabel("Nivel de contaminación")
plt.ylabel("Frecuencia")
plt.show()
```

El histograma muestra que los niveles de contaminación del agua se distribuyen de manera relativamente uniforme entre los distintos rangos (0 a 9 ppm), sin concentrarse excesivamente en un solo nivel. Esto indica que varias regiones presentan distintos grados de contaminación, lo que sugiere la necesidad de analizar cómo estos niveles afectan la incidencia de enfermedades hídricas en el siguiente paso bivariado.

```python
#gráfico bivariado - contaminación vs casos de colera
if 'cholera_cases_per_100_000_people' in df.columns and 'contaminant_level_ppm' in df.columns:
    plt.figure(figsize=(8,5))
    sns.scatterplot(data=df, x='contaminant_level_ppm', y='cholera_cases_per_100_000_people',
                    hue=df['pais'] if 'pais' in df.columns else None, alpha=0.6)
    plt.title("Nivel de contaminación vs casos de cólera")
    plt.xlabel("Nivel de contaminación")
    plt.ylabel("Casos de cólera")
    plt.show()

    # Tendencia global
    plt.figure(figsize=(8,5))
    sns.regplot(data=df, x='contaminant_level_ppm', y='cholera_cases_per_100_000_people', scatter=False, ci=None, color='red')
    plt.title("Tendencia global: contaminación → cólera")
    plt.xlabel("Nivel de contaminación")
    plt.ylabel("Casos de cólera")
    plt.show()
else:
   print("No se encontró la columna 'colera' o 'nivel_contaminacion' en el dataset.")
```

Aunque los puntos individuales están dispersos, la línea de tendencia muestra que existe una relación positiva entre el nivel de contaminación y los casos de cólera. Esto respalda la hipótesis H1: mayores niveles de contaminación se asocian con un aumento en la incidencia de enfermedades hídricas. La dispersión sugiere que otros factores (por ejemplo, acceso a agua potable, tratamiento del agua o factores socioeconómicos) también influyen en los casos reportados.

```python
#gráfico bivariado: nivel e contaminación vs tratamiento del agua

if {'water_treatment_method','contaminant_level_ppm'}.issubset(df.columns):
    plt.figure(figsize=(10,5))
    sns.boxplot(data=df, x='contaminant_level_ppm', y='water_treatment_method', palette='Set2')
    plt.title("Nivel de contaminación según método de tratamiento del agua")
    plt.xlabel("Nivel de contaminación")
    plt.ylabel("Tratamiento del agua")
    plt.xticks(rotation=45)
    plt.show()


else:
    print("No se encontró la columna 'tratamiento_agua' o 'nivel_contaminacion'.")
```

El gráfico muestra que, en promedio, los niveles de contaminación son un poco menores cuando se aplica cloración, aunque la diferencia no es muy marcada. Tanto las fuentes con tratamiento como las que no lo tienen presentan niveles significativos de contaminación, lo que sugiere que el tratamiento actual no siempre es suficiente para reducir los contaminantes. A futuro, sería interesante evaluar también la relación entre fuente de agua y nivel de contaminación, para identificar si ciertas fuentes son más críticas que otras.

Lo que muestra el gráfico es la existencia de una amplia variedad de niveles de contaminación del agua en todos los países analizados. En todos ellos, se observa que la proporción de agua no apta para consumo es considerablemente mayor que la de agua potable disponible. Aunque algunos países presentan una mayor proporción de agua no contaminada en comparación con otros, la diferencia no es significativa. **Feature selection**

```python
y = df["contaminant_level_ppm"]
X = df.drop("contaminant_level_ppm", axis=1)
```

```python
num_cols = X.select_dtypes(include=["int64", "float64"]).columns
cat_cols = X.select_dtypes(include=["object", "category"]).columns

print("Variables numéricas:", num_cols)
print("Variables categóricas:", cat_cols)
```

```python
X_encoded = pd.get_dummies(X, columns=cat_cols, drop_first=True)
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

selector = SelectKBest(score_func=f_regression, k=5)
X_new = selector.fit_transform(X_encoded, y)  # <--- y es tu variable objetivo

selected_features = X_encoded.columns[selector.get_support()]
print("Variables seleccionadas:", selected_features)

X_selected = X_encoded[selected_features]
```

```python
print("Modelo simple: Regresión lineal")
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(
    X_selected, y, test_size=0.2, random_state=42
)


from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)
y_pred_linear = linear_model.predict(X_test)
```

```python
print("modelo complejo - random forest")
```

```python
from sklearn.ensemble import RandomForestRegressor

rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

rf_model.fit(X_train, y_train)

y_pred_rf = rf_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluar_modelo(y_true, y_pred, nombre_modelo):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    print(f"----- {nombre_modelo} -----")
    print(f"MAE: {mae:.2f}")
    print(f"MSE: {mse:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"R²: {r2:.2f}\n")

# Evaluar Linear Regression
evaluar_modelo(y_test, y_pred_linear, "Linear Regression")

# Evaluar Random Forest
evaluar_modelo(y_test, y_pred_rf, "Random Forest Regressor")
```

Se entrenaron dos modelos de regresión para predecir la variable contaminant_level_ppm:
   - **Linear Regression**
   - **Random Forest Regressor**

2. Las métricas obtenidas fueron:

| Modelo | MAE | MSE | RMSE | R² |
|--------|-----|-----|------|----|
| Linear Regression | 0.54 | 0.43 | 0.65 | 0.95 |
| Random Forest Regressor | 0.60 | 0.58 | 0.76 | 0.93 |

3. Interpretación:
   - Ambos modelos presentan **errores bajos** y **R² muy alto**, indicando que pueden explicar la mayor parte de la variabilidad de la variable objetivo.
   - Linear Regression tiene un rendimiento ligeramente mejor, sugiriendo que la relación entre las variables seleccionadas y el target es **principalmente lineal**.

4. Observaciones y mejoras:
   - Los modelos funcionan muy bien con estas features; Random Forest podría ser útil si se exploraran relaciones no lineales más complejas.
   - Para un análisis más robusto, se podría realizar **cross-validation** o probar otras técnicas de regularización.
